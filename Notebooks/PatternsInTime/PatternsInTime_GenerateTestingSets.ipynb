## PatternInTime: Generating Testing Data Sets

This notebook is dedicated to generating sets of LLIF domain-like documents. It was utilized to generate datasets such as "headache," which exhibits a Monday-heavy pattern, or the BloodPressure dataset, which also shows a Monday-heavy distribution. The code within can be further expanded to create additional datasets as needed.

```python
from datetime import datetime, timedelta
import random
import json
```

```python
name = "headache"
diary_events_documents = []
starting_datetime = datetime(year=2023, month=9, day=1, hour=15)
current_datetime = starting_datetime

for i in range(160):
    if current_datetime.strftime("%A") == "Monday" or random.randint(1, 5) == 1:
        diary_events_documents.append({"name": name, "timestamp": current_datetime.isoformat()})
    
    current_datetime += timedelta(days=1) 
        
# File path to save the JSON file
file_path = "data/diary_events_monday_heavy.json"

# Writing the list of dictionaries to a JSON file
with open(file_path, "w") as json_file:
    json.dump(diary_events_documents, json_file, indent=4)

print("Data saved to", file_path)
```

```python
blood_pressure_documents = []
current_datetime = datetime(year=2024, month=1, day=1, hour=15)

for i in range(80):
    if current_datetime.strftime("%A") in ["Monday"]:
        blood_pressure_documents.append({"diastolic_blood_pressure":random.randint(60,70), "timestamp":current_datetime.isoformat()})
        current_datetime = datetime(year=2024, month=1, day=1, hour=random.randint(1,23)) + timedelta(days=i+1)
        continue
    
    if random.randint(1,3) == 1:
        blood_pressure_documents.append({"diastolic_blood_pressure":random.randint(50,60), "timestamp":current_datetime.isoformat()})
        current_datetime = datetime(year=2024, month=1, day=1, hour=random.randint(1,23)) + timedelta(days=i+1)
        
        
# File path to save the JSON file
file_path = "data/blood_pressure_diastolic_higher_on_monday.json"

# Writing the list of dictionaries to a JSON file
with open(file_path, "w") as json_file:
    json.dump(blood_pressure_documents, json_file, indent=4)

print("Data saved to", file_path)
```
